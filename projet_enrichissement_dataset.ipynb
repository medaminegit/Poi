### code insee - code postal - département - region 

```python
info_comm = pd.read_csv("https://public.opendatasoft.com/explore/dataset/correspondance-code-insee-code-postal/download/?format=csv&timezone=Europe/Berlin&lang=fr&use_labels_for_header=true&csv_separator=%3B", sep = ";")
```

```python
info_comm.head(10)
```

```python
pd.set_option('display.max_rows', None)
```

```python
info_comm["Commune"] = info_comm["Commune"].replace('\-',' ',regex=True).astype(str)
```

```python
info_comm[['Latitude', 'Longitude']] = info_comm['geo_point_2d'].str.strip('( )').str.split(',', expand=True)                  
```

```python
info_comm.sort_values(by=['Commune'], ascending=False).head(10)
```

```python
info_comm = info_comm[['Code INSEE', 'Code Postal','Latitude', 'Longitude', 'Commune', 'Département', 'Région', 'Code Département', 'Code Région']]
```

```python
info_comm[info_comm["Commune"] == "VANDOEUVRE LES NANCY"]
```

```python
info_comm.isnull().sum()
```

```python
len(info_comm["Code INSEE"].unique())
```

```python
len(info_comm)
```

### REGION
